```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ecg_data = pd.read_csv('/content/ECG_Group 4_Data 4.csv')
```

image.png

1. P time and amplitude
2. QRS time and amplitude
3. T time and amplitude
4. PR interval (q_start -p_start)
5. QT interval (T_end - q start)
6. RR interval (R_next - R_current)
7. heart rate mean

```python
def get_data(start, end, wave):
  filtered_data = ecg_data[(ecg_data['Time (sec)'] >= start) & (ecg_data['Time (sec)'] <= end)]
  # Extract the relevant columns
  time = filtered_data['Time (sec)']
  ecg_signal = filtered_data['CH 1, ECG (mV)']
  heart_rate = filtered_data['CH40, Heart Rate (mV)']

# Plot the ECG signal and Heart Rate on the same figure
  fig, ax1 = plt.subplots(figsize=(10, 6))

  # Plot ECG Signal on primary y-axis
  ax1.set_xlabel('Time (sec)')
  ax1.set_ylabel('ECG (mV)', color='b')
  ax1.plot(time, ecg_signal, color='b', label='ECG Signal')
  ax1.tick_params(axis='y', labelcolor='b')

  # Create a secondary y-axis for Heart Rate
  ax2 = ax1.twinx()
  ax2.set_ylabel('Heart Rate (BPM)', color='r')
  ax2.plot(time, heart_rate, color='r', linestyle='--', label='Heart Rate')
  ax2.tick_params(axis='y', labelcolor='r')

  # Adding titles and legend
  plt.title('ECG Signal and Heart Rate vs Time (10s - 20s)')
  fig.tight_layout()
  plt.show()
  duratition = (time.max() - time.min())* 1000
  amplitude = ecg_signal.max()
  start_ = time.min() * 1000
  end_ = time.max() * 1000
  print(f'duration= {duratition}')
  print(f'amplitude= {amplitude}')
  print(f'heart_rate= {heart_rate.mean()}')
  print(f'{wave} start= {start_}')
  print(f'{wave} end= {end_}')
  return start_, end_
```
